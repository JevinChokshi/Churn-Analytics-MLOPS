```python
dfs = {
    "app_user_accounts":app_user_accounts, 
       "app_user_profiles":app_user_profiles,
       "application_error_logs":application_error_logs,
       "auth_login_events":auth_login_events,
       "auth_session_tokens":auth_session_tokens,
       "billing_invoices":billing_invoices,
       "clickstream_events":clickstream_events,
       "experiment_assignments":experiment_assignments,
       "feature_flag_exposures":feature_flag_exposures,
       "frontend_performance_events":frontend_performance_events,
       "mobile_application_events":mobile_application_events,
       "notification_delivery_events":notification_delivery_events,
       "payment_transactions":payment_transactions,
       "product_sessions":product_sessions,
       "subscription_accounts":subscription_accounts,
       "subscription_state_history":subscription_state_history,
       "support_ticket_events":support_ticket_events,
       "support_tickets":support_tickets,
       "user_feedback_submissions":user_feedback_submissions,
       "user_device_profiles": user_device_profiles,
       "user_master_profiles": user_master_profiles}
```

```python
for name, df in dfs.items():
    print(name, ': ', df.duplicated().sum())
```

```python
BASE_DIR = Path().cwd().parents[0]
YAML_PATH = BASE_DIR / "config.yaml"

def _load_yaml() -> Dict[str, Any]:
    if not YAML_PATH.exists():
        return {}
    with open(YAML_PATH, "r", encoding="utf-8") as f:
        data = yaml.safe_load(f) or {}
    return data
```

```python
cnfg_yaml = _load_yaml()
cnfg = cnfg_yaml['database']
username = cnfg['username']
password = quote_plus(cnfg['password'])
host = cnfg['host']
port = cnfg['port']
database = cnfg['database']  

engine = sqlalchemy.create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)
# Specify your target schema name here
target_schema = "stg"

# Create the schema if it does not exist
with engine.begin() as connection:
    connection.execute(sqlalchemy.text(f"CREATE SCHEMA IF NOT EXISTS {target_schema};"))
    print(f"📁 Schema '{target_schema}' verified/created.")

# Upload DataFrames
for name, df in dfs.items():
    df.head().to_sql(
        name=f"{name}", 
        con=engine, 
        schema=target_schema, 
        index=False,
        if_exists="replace",
        
    )
    print(f"✅ Successfully uploaded {name} to schema '{target_schema}'!")
```

```python
with engine.begin() as connection:
    for name, df in dfs.items():
    
        connection.execute(sqlalchemy.text(f"TRUNCATE TABLE {target_schema}.{name};"))
        print(f"{name} Table truncated.")
```
